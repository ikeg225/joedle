```python
from nltk.corpus import words, wordnet

set_words = set(words.words())
wordnet_set = set(wordnet.words())
```

```python
all_words = set.union(set_words, wordnet_set)
```

```python
length_words = [i for i in all_words if len(i) == 6]
```

```python
print(length_words)
```
